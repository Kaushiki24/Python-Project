# 1.Create a line graph for profit and profit margin for different years:

```python
from pandas import*
import pandas as pd
from matplotlib import*
import matplotlib.pyplot as plt
from seaborn import*
import seaborn as sns
from numpy import*
import numpy as np
plt.style.use('ggplot')
```

```python
data1 = read_excel("Profit.xlsx")
data1.head()
```

```python
plt.plot(data1.Year, data1.Profit ,color ="b" ,linestyle ="--", marker = "o",label= "Profit")
plt.title("Profit vs Year")
plt.xlabel("Year")
plt.ylabel("Profit")
plt.legend()
plt.show()
```

```python
plt.plot(data1.Year, data1.Profit_Margin, color="r", linestyle="--", marker="o",label ="Profit_Margin")
plt.title("Profit_Margin vs Year")
plt.xlabel("Year")
plt.ylabel("Profit_Margin")
plt.legend()
plt.show()
```

```python
#width= 0.5
plt.subplot(2,1,1)
plt.plot(data1.Year, data1.Profit, color="b", linestyle="--", marker="o",label= "Profit", linewidth = 3)
plt.xlabel("Year")
plt.ylabel("Profit")
plt.legend()

plt.twinx()
plt.plot(data1.Year, data1.Profit_Margin , color="r", linestyle="--", marker = "*",label="Profit_Margin", linewidth = 3)
plt.xlabel("Year")
plt.ylabel("Profit_Margin")
plt.legend()
plt.suptitle("Profit/Profit_Margin vs Year")
#plt. figure(figsize=(10,8))
plt.show()
```

# 3. Pie chart for expense breakup.

```python
data3 = read_excel("Expense Breakup.xlsx")
data3.head()
```

```python
label=data3["Costs"]
#Seaborn color palette to plot pie chart
colors = sns.color_palette('bright')
#create pie chart using matplotlib
plt.pie(data3.Value, labels=label, colors=colors,autopct='%.0f%%', shadow=True, rotatelabels='true')
plt.title('Expense Breakup ')
plt.show()
```

# 4. Main expenditure item Target vs achieved

```python
data4= read_excel("Expenditure.xlsx")
data4.head()
```

```python
#width = 0.5
plt.subplot(2,1,1)
plt.plot(data4.Expenditure, data4.Target, color="b", linestyle="--", marker="o",label= "Target", linewidth = 3)
plt.xlabel("Expenditure")
plt.ylabel("Target")
plt.legend()
#plt. figure(figsize=(10,6))
plt.twinx()
plt.plot(data4.Expenditure, data4.Achieved, color="g", linestyle="--", marker = "*",label ="Achieved", linewidth = 3)
plt.xlabel("Expenditure")
plt.ylabel("Achieved")
plt.legend()
plt.suptitle("Target/Achieved vs Expenditure ")
#plt. figure(figsize=(10,8))
plt.show()
```
