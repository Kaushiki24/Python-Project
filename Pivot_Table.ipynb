# Problem Statement_1
We have data of product sales in the 'Dataset' sheet. Below are your tasks										
										
1	Draw a Pivot table to get the different types of outlets and their counts									
2	Draw a Pivot table to get the total sales for different outlets									
3	Draw a Pivot table to find out the total sales of different item types in different outlet types									
4	In the third Pivot table, filter out (remove) low fat products 																		


```python
from pandas import*
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
Sales = read_excel("Sales.xlsx")
Sales
```

# 1. Draw a Pivot table to get the different types of outlets and their counts.		

```python
outlets_type = Sales['Outlet_Type'].value_counts()
outlets_type
```

# 2. Draw a Pivot table to get the total sales for different outlets.							

```python
total_sales = pd.pivot_table(data = Sales, index=['Outlet_Type'], values = ['Item_Outlet_Sales'], aggfunc ='sum')
total_sales
```

# 3.  Draw a Pivot table to find out the total sales of different item types in different outlet types.									

```python
item_types = pd.pivot_table(data = Sales, index=['Outlet_Type','Item_Type'], values = ['Item_Outlet_Sales'], aggfunc ='sum')
item_types
```

# 4. In the third Pivot table, filter out (remove) low fat products. 							

```python
condition = ['Regular']
#rows which have not contain low fat products
filter = Sales[Sales['Item_Fat_Content'].isin(condition)]
filter
```

```python
dataset = pd.pivot_table(data = filter, index=['Outlet_Type','Item_Type','Item_Fat_Content'], values = ['Item_Outlet_Sales'], aggfunc ='sum')
dataset
```
